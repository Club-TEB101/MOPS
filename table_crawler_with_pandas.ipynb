```python
import pandas
url = 'https://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID=1104&SYEAR=2020&SSEASON=1&REPORT_ID=C'
df = pandas.read_html(url)
```

```python
import re

Balancesheet = df[0]
Balancesheet = Balancesheet.dropna()
Balancesheet = Balancesheet.iloc[:,0:2]
# sid = url.split('&')[1].split('=')[1]
# year = url.split('&')[2].split('=')[1]
# Q = url.split('&')[3].split('=')[1]
# season = f'{year}_Q{Q}'
# Balancesheet.columns = ['code','Accounting_title', 'value']
# Balancesheet = Balancesheet.drop('Accounting_title', axis = 1) # axis = 1 刪除欄位
# Balancesheet['sID'] = sid
# Balancesheet['season'] = season
# Balancesheet = Balancesheet[['sID', 'season', 'code', 'value']]
#Balancesheet['en'] = Balancesheet.iloc[:,1].apply(lambda x: x = re.search(r"[]))
Balancesheet['accounting_title']= Balancesheet.iloc[:,1]

#print(Balancesheet['accounting_title'])
```

```python
en = []
cht = []
for i in Balancesheet['accounting_title']:
    en_1 = "".join(j for j in i if ord(j) < 256)
    en.append(en_1)
    cht_1 = "".join(j for j in i if ord(j) > 256)
    cht.append(cht_1)
#print(en, cht)
```

```python
import re
eni = []
for i in en:
    j = re.sub(r'^ ', '',i)
    eni.append(j)
#print(eni)

Balancesheet['en'] = eni
#Balancesheet
```

```python
cht_1 = list(map(lambda x: x.strip().replace(u'\u3000', u''),cht))
cht_1
Balancesheet['cht'] = cht_1
Balancesheet.columns = ['code','0','1','en','cht']
Balancesheet = Balancesheet.drop(['0','1'], axis = 1)
Balancesheet
```
