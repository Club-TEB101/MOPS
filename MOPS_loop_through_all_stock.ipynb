```python
import pymysql

# https://www.itread01.com/content/1543496702.html 參考資料
# https://bigdatafinance.tw/index.php/tech/data-processing/500-python-mysql

def read_data(sql):
    host = '120.97.27.113'  #ip
    port = 3306  #埠號
    user = 'teb101Club'   #資料庫使用者名稱
    password = 'teb101Club'    #資料庫密碼
    db = 'twstock'    #資料庫名
    #charset = 'utf8mb4'  #編碼

    try:
        #db = pymysql.connect(host=host, port=port, user=user, password=password, db=db, charset=charset)  #開啟資料庫連線
        db = pymysql.connect(host=host, port=port, user=user, password=password, db=db)  #開啟資料庫連線
        cursor = db.cursor()  #建立遊標物件

        line = cursor.execute(sql)   #執行sql,返回行數
        data = cursor.fetchall()   #fetchall返回查詢的全部資料， 返回元組型別資料
        # data = cursor.fetchone()   #fetchone只獲取一條資料， 返回元組型別資料
        # print(line)
        return data
        # print(data)

    except Exception as msg:
        print(msg)
    finally:
        cursor.close()  #關掉遊標
        db.close()  #關掉資料庫連線


if __name__ == '__main__':
    sql = "SELECT sID FROM stock_info"  # 資料庫語句
    sID_tuple = read_data(sql) #返回tuple
```

```python
sID_tuple = sID_tuple[:1] #縮小

header= {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
url_list = []
for sID in sID_tuple:
    print(sID[0])
    for Q in range(1,5):
        for year in range(2015,2021):
            _url='https://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID={}&SYEAR={}&SSEASON={}&REPORT_ID=C'
            if year == 2020:
                _url = _url.format(sID[0], year, 1)
                #session = request.get(url.format())
            else:
                _url = _url.format(sID[0], year, Q)
            url_list.append(_url)
```

```python
import requests
from requests.adapters import HTTPAdapter
from bs4 import BeautifulSoup as bs4

# 此爬蟲只適用2019以後的資料
for url in url_list:
    print(url)

    web_ss = requests.session()
    ss_adapter = HTTPAdapter(max_retries=3)
    web_ss.mount("https://", adapter=ss_adapter)
    res = web_ss.get(url=url, headers=header)
    res.encoding = 'big5'
    if res.status_code == 200:
        #soup = bs4(res.text, "lxml")
        soup = bs4(res.text, "html.parser")
        tables = soup.select('table')
    #print(tables)
    
    #----------------balance sheet-----------------------------------
    bal_args = {'Ntable':0, 'start':6, 'k':3}
#     Ntable = 0
#     start = 6
#     k = 3

    code = tables[bal_args['Ntable']].select('td')

    code_list=[]

    for _code in code:
        try:
            if (_code['style'])!='':
                code_list.append(_code.text)
        except:
            pass
    #print(code_list)

    #cht = tables[bal_args['Ntable']].select('span.zh')
    #en = tables[bal_args['Ntable']].select('span.en')
    amt = tables[bal_args['Ntable']].select('td.amt')

    #print(code)
    #print(len(cht[start:]), len(amt)/k)
    #df_cht, df_eng = {} , {}

    Balance_list=[]
    j=0
    for i, _amt in enumerate(amt):

        if i%bal_args['k']==0:
            #print(i, _amt.text)
            #balance_list[code_list[j]]= [{'cht':cht[6+j].text.strip(),'value':_amt.text.strip().replace(',','')}]
            #Balance_list.append({'code':code_list[j], 'en':en[bal_args['start']+j].text.strip(), 'cht':cht[bal_args['start']+j].text.strip(),'value':_amt.text.strip().replace(',','')})
            Balance_list.append({'code':code_list[j],'value':_amt.text.strip().replace(',','')})
            j+=1
    
    sid = url.split('&')[1].split('=')[1]
    year = url.split('&')[2].split('=')[1]
    Q = url.split('&')[3].split('=')[1]
    #Balance_Q={f'{year}_Q{Q}':Balance_list} 
    #Balance = {'Balancesheet':Balance_Q}
    company={'sID':f'{sid}','season': f'{year}_Q{Q}', 'Balancesheet': Balance_list}
    print(company)
    
    
```

```python
import pandas as pd
Balance_sheet = {}
Balance_sheet['sID'] = []
Balance_sheet['season'] = []
Balance_sheet['code'] = []
Balance_sheet['value'] = []

for i in company['Balancesheet']:
    Balance_sheet['sID'].append(company['sID'])
    Balance_sheet['season'].append(company['season'])
    Balance_sheet['code'].append(i['code'])
    Balance_sheet['value'].append(i['value'])




Balance_sheet = pd.DataFrame(Balance_sheet)
print(Balance_sheet)

#取出non-empty code 的 row
Balancesheet_code_table = Balance_sheet[Balance_sheet['code'] != '']
Balancesheet_code_table.to_csv('Balancesheet_code_table.csv',index = False)
```
